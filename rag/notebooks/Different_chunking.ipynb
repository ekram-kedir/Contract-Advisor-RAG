#### 4. Semantic Chunking

```python
with open('./../prompts/context.txt') as file:
    essay = file.read()
semantic_chunks = chunker_manager.semantic_chunking(essay)
```

```python
for i, chunk in enumerate(semantic_chunks[:2]):
    buffer = 200
    
    print (f"Chunk #{i}")
    print (chunk[:buffer].strip())
    print ("...")
    print (chunk[-buffer:].strip())
    print ("\n")
```
