```python
# Parse the JSON data
import json

# Load the JSON data from the file
with open('/Users/ekramkedir/Desktop/real/Contract-Advisor-RAG/data/prepare_policy_data /data/qa_with_context/train.json', 'r') as file:
    data = json.load(file)
```

```python
# Extract "qas" from the data
qas = []
for paragraph in data['data'][0]['paragraphs']:
    qas.extend(paragraph['qas'])

# Extract questions, document IDs, and answers
qas_info = []
for qa in qas:
    qas_info.append({
        'doc_id': qa['id'],
        'question': qa['question'],
        'answers': [answer['text'] for answer in qa['answers']]
    })

# Print the extracted information
for item in qas_info:
    print("Document ID:", item['doc_id'])
    print("Question:", item['question'])
    print("Answers:", item['answers'])
    print()
```
